```python
## Recoding variables

# Variable 1: sub_grade
# First, we show why this can be recoded

# Get the average default rate per sub grade
sub_grade_data = df.groupby("sub_grade")["loan_status"].mean()

# Create a figure and plot the average default rate per sub grade
plt.figure(figsize=(9, 6))
plt.plot(sub_grade_data, marker='o', linestyle='-', color = 'blue', alpha=0.4)

# Optimize, save, and present the figure
plt.xticks(rotation=90)
plt.xlabel("Loan Subgrade")
plt.ylabel("Average Default Rate")
plt.grid(linestyle="--", alpha=0.5)
plt.savefig('Default_Rate_Loan_Subgrade.png', transparent=True)
plt.show()

# Recode the sub_grade such that A1 : 1, A2 : 2, ... , A5 : 5, B1 : 6, ... , G5 : 35
category_mapping = {f"{letter}{num}": i for i, (letter, num) in enumerate([(l, n) for l in "ABCDEFG" for n in range(1, 6)], start=1)}
df['sub_grade_recode'] = df['sub_grade'].map(category_mapping)

# Variable 2: home_ownership

# Recode the home ownership status such that "MORTGAGE" : 0, "RENT" : 2, everything else : 1
home_ownership_mapping = {"MORTGAGE" : 0, "RENT": 2}
df["home_ownership_recode"] = df["home_ownership"].map(home_ownership_mapping).fillna(1).astype(int)

# Variable 3: verification_status

# Recode the verification status such that "not verified" : 1, everything else : 0
df["verification_status_recode"] = (df["verification_status"] == "Not Verified").astype(int)

# Variable 4: issue_d

# Recode the sub_grade such that only the year is noted
df['issue_d_recode'] = pd.to_datetime(df['issue_d']).dt.year

## Transforming variables

# Variable 1: annual_inc
df["annual_inc_transf"] = np.log(df["annual_inc"])
```

```python
# Function to compute the information value given the full data of independent and dependent variable
def compute_iv(df, exo_var):
    
    # There are four exogenous variables that have at least one value missing, so drop missing values if needed
    df_cleaned = df.dropna(subset=[exo_var]).copy()
    
    # Check if value is not a String and if value has a unique value with less than 20 observations
    check_a = not df[exo_var].apply(lambda x: isinstance(x, str)).any()
    check_b = (df[exo_var].value_counts() < 20).any()
    
    # If check_a and check_b are are both True, apply binning
    if (check_a and check_b):
        df_cleaned.loc[:,exo_var] = pd.qcut(df_cleaned[exo_var], q=bins_amount, duplicates='drop')

    # Define the dependent variable
    dep_var = "loan_status"

    # Algorithm to compute the information value
    grouped_info = df_cleaned.groupby(exo_var)[dep_var].agg(['count','sum'])
    grouped_info = grouped_info[grouped_info['count']>=5]

    nBads = grouped_info['sum']
    nGoods = grouped_info['count'] - grouped_info['sum']
    
    per_Bad = nBads / np.sum(nBads)
    per_Good = nGoods / np.sum(nGoods)
   
    IV = np.sum((per_Bad - per_Good) * np.log(per_Bad / per_Good))
    
    # Return the information value
    return IV
```

```python
## Compute the information value per variable

# Loop over each exogenous variable and if it has at least one missing data point, introduce a new binary variable
for exo_var in (df.columns[2:]):
    if (np.sum(df[exo_var].isna()) > 0):
        df.loc[:,exo_var+"_NaN_bin"] = df[exo_var].isna().astype(int)

# Create an empty list to store results of information values
iv_results = []

# Loop over each exogenous variable and compute the information value per variable
for exo_var in (df.columns[2:]):
    iv_value = compute_iv(df, exo_var)
    
    # Check which type the data of the variable is
    var_type = "numeric" if pd.api.types.is_numeric_dtype(df[exo_var]) else "text"
    
    # Store the information value in the list
    iv_results.append({"Variable": exo_var, "Information Value": iv_value, "Variable Type": var_type})

# Convert the list into a DataFrame
iv_df = pd.DataFrame(iv_results)

# Sort the DataFrame by information value
iv_df = iv_df.sort_values(by="Information Value", ascending=False).reset_index(drop=True)

# Display the DataFrame
iv_df
```

```python
## Splitting data into training and test data

# Seperating the data into a the independent and dependent variables
Ind_vars = df.iloc[:,2:]
Dep_var = df["loan_status"]

# Split the data into 80% training and 20% test data
X_train, X_test, y_train, y_test = train_test_split(Ind_vars, Dep_var, test_size=0.2, random_state=1, shuffle=True)

# Compute column-wise mean from X_train
train_means = X_train.mean()

# Fill missing values in both X_train and X_test with the respective column mean from X_train
X_train = X_train.copy()
X_test = X_test.copy()
X_train.fillna(train_means, inplace=True)
X_test.fillna(train_means, inplace=True)
```

```python
# More imports
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer, StandardScaler
from sklearn.linear_model import LogisticRegression

# Set a fixed seed
np.random.seed(1)

# Select the relevant variables
selected_variables = iv_df.loc[(iv_df["Variable Type"] == "numeric") & (iv_df["Information Value"] >= 0.02) & ~iv_df["Variable"].isin(["int_rate","fico_range_high","installment",'annual_inc']), "Variable"].values

# Create the custom polynomial
def custom_poly(X):
    X_new = np.hstack([
        # Function 'age'
        X.iloc[:, [0]],
        X.iloc[:, [0]] ** 0.5,
        X.iloc[:, [0]] ** 2,
        X.iloc[:, [0]] ** 3,
        np.log(X.iloc[:, [0]]),
        np.exp(X.iloc[:, [0]]),
        # Function 'sub_grade_recode'
        X.iloc[:, [1]],
        X.iloc[:, [1]] ** 0.5,
        X.iloc[:, [1]] ** 2,
        X.iloc[:, [1]] ** 3,
        # Function 'term'
        X.iloc[:, [2]],
        # Function 'pay_status'
        X.iloc[:, [3]],
        X.iloc[:, [3]] ** 2,
        X.iloc[:, [3]] ** 3,
        # Function 'fico_range_low'
        X.iloc[:, [4]],
        X.iloc[:, [4]] ** 0.5,
        X.iloc[:, [4]] ** 2,
        X.iloc[:, [4]] ** 3,
        # Function 'dti'
        X.iloc[:, [5]],
        X.iloc[:, [5]] ** 0.5,
        X.iloc[:, [5]] ** 2,
        X.iloc[:, [5]] ** 3,
        # Function 'verification_status_recode'
        X.iloc[:, [6]],
        # Function 'annual_inc_transf'
        X.iloc[:, [7]],
        X.iloc[:, [7]] ** 0.5,
        X.iloc[:, [7]] ** 2,
        X.iloc[:, [7]] ** 3,
        # Function 'mort_acc'
        X.iloc[:, [8]],
        X.iloc[:, [8]] ** 0.5,
        X.iloc[:, [8]] ** 2,
        X.iloc[:, [8]] ** 3,
        # Function 'loan_amnt'
        X.iloc[:, [9]],
        X.iloc[:, [9]] ** 0.5,
        X.iloc[:, [9]] ** 2,
        X.iloc[:, [9]] ** 3,
        # Function 'home_ownership_recode'
        X.iloc[:, [10]],
        X.iloc[:, [10]] ** 0.5,
        X.iloc[:, [10]] ** 2,
        X.iloc[:, [10]] ** 3,
        # Function 'issue_d_recode'
        X.iloc[:, [11]],
        X.iloc[:, [11]] ** 0.5,
        X.iloc[:, [11]] ** 2,
        X.iloc[:, [11]] ** 3,
        # Function 'revol_util'
        X.iloc[:, [12]],
        X.iloc[:, [12]] ** 0.5,
        X.iloc[:, [12]] ** 2,
        X.iloc[:, [12]] ** 3,
    ])
    return X_new

# Create polynomial logistic regression with LASSO regularization
model = Pipeline([
    ('poly', FunctionTransformer(custom_poly)),  # Second-degree polynomial terms
    ('scaler', StandardScaler()),  # Normalize features for better regularization
    ('logreg', LogisticRegression(penalty='l1', solver='liblinear', C=1, class_weight='balanced',max_iter=1000))  # LASSO regularization
])

# Fit model
model.fit(X_train[selected_variables], y_train)
```

```python
# More imports
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix
from sklearn.metrics import log_loss, roc_auc_score, roc_curve

# Print coefficients (Intercept and all other parameters)
print("Intercept:", model.named_steps['logreg'].intercept_)
print("Coefficients:", np.round(model.named_steps['logreg'].coef_,5))
print()

# Print the confusion matrix
print("The confusion Matrix")
print(confusion_matrix(y_test, model.predict(X_test[selected_variables])))
print()

# Get predicted probabilities
y_prob = model.predict_proba(X_test[selected_variables])[:, 1]

# Compute Gini coefficient
gini_coef = 2 * roc_auc_score(y_test, y_prob) - 1

print("Gini Coefficient: " + str(np.round(gini_coef,4)))

# 6. ROC Curve (to visualize model performance)
fpr, tpr, _ = roc_curve(y_test, y_prob)
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label='Logistic regression Model', color='maroon', linewidth = 1.2, alpha=1)
plt.plot([0, 1], [0, 1], linestyle='--', label='Random classifier', color = 'blue', alpha=0.4)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.legend(loc="lower center", bbox_to_anchor=(0.5, 1), ncol=2,framealpha=0.1)
plt.savefig('ROC_curve.png', transparent=True)
plt.show()
```
